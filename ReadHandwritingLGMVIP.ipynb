```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

IMPORTING DATASET

```python
data_set= tf.keras.datasets.mnist
```

CONVERT DATA INTO X_TRAIN & Y_TRAIN

```python
(x_train,y_train),(x_test,y_test)=data_set.load_data()
```

```python
fig=plt.figure(figsize=(13,3))
for i in range(16):
  ax=fig.add_subplot(2,8,i+1)
  ax.imshow(np.squeeze(x_train[i]),cmap='Reds')
  ax.set_title(y_train[i])
```

```python
print(x_train.shape)
print(x_train[0])
```

```python
xtrain = x_train/255.0
xtest = x_test/255.0
```

```python
data_model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation ='relu'),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
data_model.summary()
```

ADAM OPTIMIZER

```python
data_model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss ='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

CALCULATING THE ACCURACY AND EXAMINING DATA INTO 'N' REPETITIONS

```python
data_model.fit(xtrain,y_train, epochs=8)
```

```python
data_model.fit(xtrain,y_train, epochs=10)
```

```python
history=data_model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.1)
```

PLOTS OF ACCURACY X LOSS MODEL

```python
plt.title("Loss X Accuracy")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history["val_accuracy"],label='val')
plt.show()
```

```python
plt.title("Loss X Accuracy")
plt.xlabel("Epoch")
plt.ylabel("acc/Loss")
plt.plot(history.history['loss'],label='accuracy')
plt.plot(history.history["val_loss"],label='val')
plt.show()
```
